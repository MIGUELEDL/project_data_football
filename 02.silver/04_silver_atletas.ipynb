```python
from pyspark.sql.functions import col
```

```python
df_atleta = spark.table("project_data_football_bronze.atletas")
```

```python
%sql
describe project_data_football_bronze.atletas
```

```python
df_dim_atleta = df_atleta.select(
    col("atleta_id").cast("int").alias("atleta_id"),
    col("nome"),
    col("apelido"),
    col("clube_id").cast("int"),
    col("posicao_id").cast("int"),
    col("status_id").cast("int"),
    col("dt_ingestao").alias("dt_ultima_atualizacao")
).dropDuplicates(["atleta_id"])
```

```python
display(df_dim_atleta)
```

```python
df_dim_atleta.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("project_data_football_silver.dim_atleta")
```
