```python
from pyspark.sql.functions import col, current_timestamp, lit
```

```python
df_partidas = spark.table("project_data_football_bronze.partidas_rodada")
```

```python
%sql
describe project_data_football_bronze.partidas_rodada
```

```python
df_dim_campeonato = (
    df_partidas
        .select(col("campeonato_id").cast("int"))
        .dropDuplicates(["campeonato_id"])
        .withColumn("nome", lit("Brasileirão Série A"))
        .withColumn("temporada", current_timestamp().cast("date").substr(1,4).cast("int"))
        .withColumn("dt_ultima_atualizacao", current_timestamp())
)
```

```python
display(df_dim_campeonato)
```

```python
df_dim_campeonato.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("project_data_football_silver.dim_campeonato")
```
